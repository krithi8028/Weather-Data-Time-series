```python
import numpy as np
import re
import pandas as pd
df=pd.read_csv('Weather_chennai_1.csv',delimiter=',')
```

```python
data=pd.DataFrame(df)
```

```python
df[['day', 'time_range']] = df['date'].str.split(',', expand=True)
df['time_range'] = df['time_range'].str.strip()
```

```python
df
```

```python
df['day'] = pd.to_datetime(df['day'], format='%A %d %B %Y')
df['month'] = df['day'].dt.month
```

```python
df['day_date']=df['day'].dt.day
```

```python
df['year']=df['day'].dt.year
df['day_name']=df['day'].dt.day_name()
```

```python
df.head()
```

```python
df[['Start_hour', 'End_hour']] = df['time_range'].str.split('—', expand=True)

df['Start_hour'] = df['Start_hour'].str.strip()
df['End_hour'] = df['End_hour'].str.strip()

```

```python
df
```

```python
df=df.drop(('time_range'),axis=1)
```

```python
df.columns
```

```python
df=df[['date', 'date_id', 'wind', 'desc', 'temp', 'baro', 'hum',
       'day', 'month', 'day_date', 'year','day_name', 'Start_hour', 'End_hour']]
```

```python
df[['wind','temp','baro']]=df[['wind','temp','baro']].astype(float)
df[['hum','year','day_date']]=df[['hum','year','day_date']].astype(int)
```

```python
from datetime import datetime
df['Start_hour'] = pd.to_datetime(df['Start_hour'], format='%H.%M').dt.strftime('%H:%M')
df['End_hour'] = pd.to_datetime(df['End_hour'], format='%H.%M').dt.strftime('%H:%M')
```

```python
df
```

```python
days={'Monday':0, 'Tuesday':1,'Wednesday': 2, 'Thursday':3, 'Friday':4,'Saturday': 5, 'Sunday':6}
df['day_name']=df['day_name'].map(days)
df.head(2)
```

```python
df.to_csv('Clean15Year.csv',index=False)
```
