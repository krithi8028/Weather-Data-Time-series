```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('Clean15YearChennaiWeather.csv')
```

```python
df1 = df.groupby(['month'])['temp'].mean().reset_index()
df2 = df.groupby(['month'])['hum'].mean().reset_index()

ax = df1.plot(x='month', y='temp', kind='line', color='green')
ax2 = df2.plot(x='month', y='hum', kind='line', color='red', ax=ax, secondary_y=True)

ax.set_ylabel("Temperature")
ax.right_ax.set_ylabel("Humidity")
ax.set_xlabel("Month")
```
