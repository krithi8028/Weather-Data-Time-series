```python
month=range(1,13)
year=range(2010, 2024)
def insertchar(mystring,month,year):
    mystring=mystring[:65]+year+mystring[65:71]+month+mystring[71:]
    return mystring
result_url=[]
for i in year:
    for j in month:
        result_url.append(insertchar(url,str(i),str(j)))
        
result_url
```

```python
import json
import os
def extract_weather_data(file_path):
    with open(file_path, encoding='utf-8') as f:
        r = f.read()
    soup = BeautifulSoup(r, "html.parser")
    scripts = soup.find_all('script')
    for script in scripts:
        if 'var data=' in script.text:
            jsonStr = script.text.strip().split('var data=')[1].split(';')[0]
            return json.loads(jsonStr)
    return None

def process_json_data(jsonObj):
    if not jsonObj:
        return []
    return [
        {
            'date': item.get('ds'),
            'date_id': item.get('date'),
            'wind': item.get('wind'),
            'desc': item.get('desc'),
            'temp': item.get('temp'),
            'baro': item.get('baro'),
            'hum': item.get('hum')
        }
        for item in jsonObj.get('detail', [])
    ]

# Directory containing the Soupfiles
directory = "C:/Users/KRITHIKA/"

# Initialize an empty DataFrame to store all data
dfs = pd.DataFrame()

# Iterate through all files in the directory
for filename in os.listdir(directory):
    if filename.startswith("Soupfile_") and filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        
        # Extract JSON data from the file
        jsonObj = extract_weather_data(file_path)
        
        # Process the JSON data
        ddf_rows = process_json_data(jsonObj)
        
        # If we have data, add it to the main DataFrame
        if ddf_rows:
            ddf = pd.DataFrame(ddf_rows, columns=['date', 'date_id', 'wind', 'desc', 'temp', 'baro', 'hum'])
            dfs = pd.concat([dfs, ddf], ignore_index=True)

        print(f"Processed {filename}")

# Print the first 5 rows and basic info about the resulting DataFrame
print(dfs.head())
print(dfs.info())
```

```python
import logging # Logging is essential for tracking the script's progress and diagnosing issues
logging.basicConfig(
    filename='weather_scraping.log',  # Log file name
    filemode='w',                     # Overwrite log file each run
    level=logging.INFO,               # Logging level
    format='%(asctime)s - %(levelname)s - %(message)s'  # Log message format
)
```

```python
# Hindi to English translations for months
hindi_to_english_months = {
    'जनवरी': 'January',
    'फरवरी': 'February',
    'मार्च': 'March',
    'अप्रैल': 'April',
    'मई': 'May',
    'जून': 'June',
    'जुलाई': 'July',
    'अगस्त': 'August',
    'सितंबर': 'September',
    'अक्टूबर': 'October',
    'नवंबर': 'November',
    'दिसंबर': 'December'
}

# Hindi to English translations for days
hindi_to_english_days = {
    'सोमवार': 'Monday',
    'मंगलवार': 'Tuesday',
    'बुधवार': 'Wednesday',
    'गुरुवार': 'Thursday',
    'शुक्रवार': 'Friday',
    'शनिवार': 'Saturday',
    'रविवार': 'Sunday'
}
```

```python
def translate_text(text):

    if not isinstance(text, str):
        return text  # Return as-is if not a string
    
    # Translate months
    for hindi, english in hindi_to_english_months.items():
        text = text.replace(hindi, english)
    
    # Translate days
    for hindi, english in hindi_to_english_days.items():
        text = text.replace(hindi, english)
    
    return text
```

```python

weather_data = []
soup_files_directory = r"C:/Users/KRITHIKA/"
soup_file_pattern = "Soupfile_{}.txt"
```

```python
# Total number of soup files
total_files = 167

for idx in range(total_files):
    file_name = soup_file_pattern.format(idx)
    file_path = os.path.join(soup_files_directory, file_name)
    
    # Check if the file exists
    if not os.path.exists(file_path):
        logging.warning(f"File not found: {file_path}")
        continue  # Skip to the next file
    
    try:
        # Read the HTML content from the file
        with open(file_path, 'r', encoding='utf-8') as file:
            html_content = file.read()
        
        # Parse the HTML with BeautifulSoup
        soup = BeautifulSoup(html_content, "lxml")  # Using 'lxml' parser for better performance
        
        # Locate the table containing weather data
        # Update 'wt-his' to the actual table id if different
        table = soup.find('table', id='wt-his')
        
        if not table:
            logging.warning(f"No data table found in file: {file_path}")
            continue  # Skip to the next file
        
        # Iterate over each row in the table body
        for row in table.tbody.find_all('tr'):
            # Extract all cells in the row
            cells = row.find_all('td')
            
            # Ensure there are enough cells to extract data
            if len(cells) < 5:
                logging.warning(f"Incomplete data row in file: {file_path}")
                continue  # Skip incomplete rows
            
            # Extract data from each cell
            try:
                # Time is within the first cell's 'th' tag
                time_cell = row.find('th')
                time = time_cell.get_text(strip=True) if time_cell else None
                
                # Weather is within the first 'td' with class 'small'
                weather_cell = row.find('td', class_="small")
                weather = weather_cell.get_text(strip=True) if weather_cell else None
                
                # Wind is within the second 'td' with class 'sep'
                wind_cells = row.find_all('td', class_="sep")
                wind = wind_cells[0].get_text(strip=True) if len(wind_cells) >= 1 else None
                
                # Humidity is within the third 'td' with class 'sep'
                humidity = wind_cells[1].get_text(strip=True) if len(wind_cells) >= 2 else None
                
                # Temperature is within the first 'td' without a class
                temp_cell = row.find('td', class_=False)
                temp = temp_cell.get_text(strip=True) if temp_cell else None
                
                # Translate Hindi terms to English in the 'time' field
                translated_time = translate_text(time)
                
                # Append the extracted and translated data to the list
                weather_data.append({
                    'File': file_name,
                    'URL': f"https://www.timeanddate.com/weather/india/chennai/historic?month=&year=",  # Placeholder URL
                    'Time': translated_time,
                    'Weather': weather,
                    'Temperature': temp,
                    'Humidity': humidity,
                    'Wind': wind
                })
                
            except Exception as e:
                logging.error(f"Error parsing row in file {file_path}: {e}")
                continue  # Skip problematic rows
        
        logging.info(f"Successfully processed file: {file_path}")
        
    except Exception as e:
        logging.error(f"Error processing file {file_path}: {e}")
        continue  # Skip problematic files
```

```python

df = pd.DataFrame(weather_data)
df = df[['File', 'URL', 'Time', 'Weather', 'Temperature', 'Humidity', 'Wind']]
print(df.head())

output_csv_path = r'C:/Users/KRITHIKA/Weather_chennai_consolidated.csv'
df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')  

print(f"Data collection complete. Saved to '{output_csv_path}'.")
```
